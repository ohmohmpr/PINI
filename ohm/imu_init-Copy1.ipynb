```python
idx = 16267
handfree_data = np.loadtxt('../dvs_imu_data.txt', delimiter=' ', unpack=True)[:, :idx]
```

```python
start_time = handfree_data[0][0]
handfree_data[0] = handfree_data[0] - start_time
```

```python
figure, axis = plt.subplots(2, 1, sharex=True, sharey=True, figsize=(14, 6))
# For angular velocity
# axis[0].plot(handfree_data[0], handfree_data[1])
# axis[0].plot(handfree_data[0], handfree_data[2])
axis[0].plot(handfree_data[0], handfree_data[4])
axis[0].set_title("angular velocity")
# For linear acceleration
# axis[1].plot(handfree_data[0], handfree_data[4])
axis[1].plot(handfree_data[0], handfree_data[5])
# axis[1].plot(handfree_data[0], handfree_data[6])
axis[1].set_title("linear acceleration")

figure.supxlabel('Time [sec]')
figure.supylabel('distance [m] in kitti format')
plt.show()
```

```python
init_acc_mean = handfree_data[4:7, :idx].mean(axis=1)
init_acc_mean
```

```python
init_roll = np.arctan2(-init_acc_mean[1], -init_acc_mean[2])
np.degrees(init_roll)
```

```python
init_pitch = np.arctan2(init_acc_mean[0], np.sqrt(init_acc_mean[1] * init_acc_mean[1] +
                                   init_acc_mean[2] * init_acc_mean[2]));
np.degrees(init_pitch)
```
