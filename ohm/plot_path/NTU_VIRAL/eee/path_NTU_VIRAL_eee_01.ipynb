```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
LIOEKF_imu_imu = np.loadtxt('../../../dataset/rosbag/ntu_viral/eee_01/lio_ekf_01.txt', usecols=(1,2,3), unpack=True)
LIOEKF_IN_PIN_imu_imu = np.loadtxt('../../../dataset/rosbag/ntu_viral/eee_01/pin_lio_ekf_odo.txt', usecols=(1,2,3), skiprows=65, unpack=True)
LIOEKF_imu_imu_2 = np.loadtxt('../../../dataset/rosbag/ntu_viral/eee_01/odo.txt', usecols=(1,2,3), skiprows=63, unpack=True)
# LIOEKF_IN_PIN_dvs_imu = np.loadtxt('../dataset/rosbag/m2dgr/street_03/handsfree_imu/pin_lio_ekf_odo_dvs.txt', usecols=(1,2,3), skiprows=63, unpack=True)
pin_odom = np.loadtxt('../../../dataset/rosbag/ntu_viral/eee_01/odom_poses_kitti.txt', delimiter=' ', usecols=(3, 7, 11), unpack=True)
pin_slam = np.loadtxt('../../../dataset/rosbag/ntu_viral/eee_01/slam_poses_kitti.txt', delimiter=' ', usecols=(3, 7, 11), unpack=True)
```

```python
trans = np.array([[ 1,  0,  0],
                  [ 0,  1,  0],
                  [ 0,  0, -1]])

LIOEKF_imu_imu = trans @ LIOEKF_imu_imu
pin_odom = trans @ pin_odom
pin_slam = trans @ pin_slam
LIOEKF_imu_imu_2 = trans @ LIOEKF_imu_imu_2

# trans = np.array([[ 1,  0,  0],
#                   [ 0,  -1,  0],
#                   [ 0,  0, -1]])
LIOEKF_IN_PIN_imu_imu = trans @ LIOEKF_IN_PIN_imu_imu
```
